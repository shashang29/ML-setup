```python
heart= pd.read_csv('heart-disease.csv')
```

```python
heart.age.plot.hist(bins=50);
```

```python
heart.plot.hist(figsize=(10,30), subplots=True);
```

```python
over50 = heart[heart['age']> 50]
over50.head()
```

```python
#pyplot method
over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target')
```

```python
# oo method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
           ax=ax);
ax.set_xlim([45,100]); #x axis limit 
```

```python

plt.style.use('seaborn-whitegrid')

# oo method fro scratch
fig, ax = plt.subplots(figsize=(10,6))
scatter= ax.scatter(x=over50['age'], y=over50['chol'],
                   c=over50['target'],
                   cmap=('Dark2') #change colour scheme
                   );
#customize;
ax.set(title='heart disease above age 50',
      xlabel='age',
      ylabel='cholestrol level')
ax.legend(*scatter.legend_elements(), title='target');
#add horizontal line
ax.axhline(over50['chol'].mean(),
          ls='--');
```

```python
# subplot of chol, age and thalach
fig, (ax0,ax1)= plt.subplots(nrows=2,ncols=1,
                             figsize=(10,10),
                             sharex=True);

scatter= ax0.scatter(x=over50['age'],
                    y=over50['chol'],
                    c=over50['target']);

ax0.set(title='heart disease above age 50',
      ylabel='cholestrol level');
ax0.legend(*scatter.legend_elements(), title='target');
ax0.axhline(y=over50['chol'].mean(),
           ls='--');

scatter=ax1.scatter(x=over50.age,
                   y=over50.thalach,
                   c=over50.target);

ax1.set(title='Heart disease and max heart rate',
       xlabel='age',
       ylabel='max heart rate');
ax1.legend(*scatter.legend_elements(), title='target');
ax1.axhline(over50.thalach.mean(),ls='--');

fig.suptitle('Heart Disease analysis', 
             fontsize=15, fontweight='bold');

```
